```python
import numpy as np
import math
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
L = 5.26
C = 0.65
R = 0.55
I = [0.0]
Q = [0.0]
V = 3.0
w = 1.4
h = 0.05
t = np.linspace(0,10,200+1)
```

```python

def f(t, Q , I):
    return I

def g(t, Q, I):
    return ((-(I*R)/L) - (Q/ (C * L)) + ((V/L) * np.sin(2 * np.pi * w * t)))
```

```python
for i in range(200):
    k1 = h * f( t[i], Q[i], I[i])
    l1 = h * g( t[i], Q[i], I[i])
    k2 = h * f( t[i] + h/2, Q[i] + k1 / 2 , I[i] + l1 / 2 )
    l2 = h * g(t[i] + h/2, Q[i] + k1/2, I[i] + l1 / 2)
    k3 = h * f(t[i] + h/2, Q[i] + k2/2, I[i] + l2 /2)
    l3 = h * g(t[i] + h/2, Q[i] + k2/2, I[i] + l2/2)
    k4 = h * f(t[i] + h, Q[i] + k3, I[i] + l3 )
    l4 = h * g(t[i] + h, Q[i] + k3, I[i] + l3)
    Q.append(Q[i] + (k1 + 2 * k2 + 2 * k3 + k4)/6)
    I.append(I[i]  + (l1 + 2 * l2 + 2 * l3 + l4)/6)
```

```python
plt.plot(t,Q)
```

```python
plt.plot(t,I)
```
