```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['Solarize_Light2'])

```

```python
vi=700
g=9.8
k=4*(10**(-5))
h=0.1
o=np.arange(30,55+5,5)
theta=np.radians(o)
t=np.arange(0,90+h, h)
vx=np.zeros((len(t),len(theta)))
vy=np.zeros((len(t),len(theta)))
x=np.zeros((len(t),len(theta)))
y=np.zeros((len(t),len(theta)))
v=np.zeros((len(t),len(theta)))
```

```python
def f(vx,vy):
    return (- k * (np.sqrt(vx**2+vy**2)) * vx)

def u(vx,vy):
    return (- g - k * (np.sqrt(vx**2+vy**2)) * vy)
```

```python
for j in range(len(theta)):
    
    vx[0,j]=vi*np.cos(theta[j])
    vy[0,j]=vi*np.sin(theta[j])
    
    for i in range(0,len(t)-1):
        
        k1=f(vx[i,j],vy[i,j])
        l1=u(vx[i,j],vy[i,j])
        
        
        k2=f(vx[i,j]+(.5*h*k1),vy[i,j]+(.5*h*l1))
        l2=u(vx[i,j]+(.5*h*k1),vy[i,j]+(.5*h*l1))

        
        k3=f(vx[i,j]+(.5*h*k2),vy[i,j]+(.5*h*l2))
        l3=u(vx[i,j]+(.5*h*k2),vy[i,j]+(.5*h*l2))
        
        k4=f(vx[i,j]+h*k3,vy[i,j]+(l3*h))
        l4=u(vx[i,j]+h*k3,vy[i,j]+(l3*h))
        
        
        vx[i+1,j]=vx[i,j]+((h/6)*(k1+2*k2+2*k3+k4))
        vy[i+1,j]=vy[i,j]+((h/6)*(l1+2*l2+2*l3+l4))     
        

        
        
for j in range(len(theta)):
    x[0,j]=0
    y[0,j]=0
    for i in range(len(t)-1):
        m1=vx[i,j]
        m2=vx[i,j]+m1*h/2
        m3=vx[i,j]+m2*h/2
        m4=vx[i,j]+m3*h
        x[i+1,j]=x[i,j]+(h/6)*(m1+2*m2+2*m3+m4)
    for i in range(len(t)-1):
        n1=vy[i,j]
        n2=vy[i,j]+n1*h/2
        n3=vy[i,j]+n2*h/2
        n4=vy[i,j]+n3*h
        y[i+1,j]=y[i,j]+(h/6)*(n1+2*n2+2*n3+n4)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1)
ax.set_ylim((0,12000))
plt.suptitle("Trajectory for $v_{initial}=700m/s $ with air resistance $ (4 * {10^{-5}m^{-1})}}$",size=20)
for j in range(len(theta)):
    plt.plot(x[:,j],y[:,j],label='$%i^\circ$'%o[j])
plt.xlabel("x",size=20)
plt.ylabel("y",size=20)    
plt.legend(loc='best',prop={'size':'x-large'})
plt.show()
```
